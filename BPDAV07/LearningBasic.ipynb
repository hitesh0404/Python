```python
from functools import lru_cache
import time
@lru_cache(maxsize=None)
def fibbonacci(n):
    if n<=1:
        return n
    else:
        time.sleep(1)
        return fibbonacci(n-1)+fibbonacci(n-2)

for i in range(10):
    print(fibbonacci(i))
```

### Arbitrary Arguments in Python
* Arbitrary arguments allow a function to accept a variable (unknown) number of arguments.
* Used when you don’t know how many positional arguments will be passed to your function.
All extra arguments are collected into a tuple.

```python
def sum(*a:tuple):
    print(a)
    result = 0
    for i in a:
        result +=i
    return result
```

```python
print(sum(0,1,2,3,4,3,2))
print(sep=" ",end=" ")
```

### Arbitrary Keyword Arguments
* Used when you don’t know how many keyword arguments will be passed.

## Comprehension

```python
data = [2,3,4,61,21,2,21]
result = []
for i in data:
    if i%2 == 0:
        result.append(f"{i} is even")
    else:
        result.append(f"{i} is odd")

print(result)
```
