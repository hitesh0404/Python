# open function
- it find the file location
- create one file object 
- we can set the mode in which we want to open the file
- default mode is read('r')

# Modes
- r read point at start and keep the file content
- w write point at start after removing the content
- a append point at last and keep the file content
- x (exclusive create) generate error if file exist works as w
- r+ read and write point at start of the file and keep the content
- w+ write and read point at start after removing the content
- a+ append and read point at last and keep the file content
- x+ (exclusive create) generate error if file exist works as w+

```python
file = open("hello1.txt",'a')
file.write("Hello world")
file.close()
```

```python
with open("hello.txt","w") as f:
    f.write("hey its a with block ")

print("file closed autometically")
f.write("done")
```

```python
with open("hello1.txt","r") as f:
    d = f.readline()
    d = f.readline()
    d = f.readline()
```

```python
import csv
```

```python
data = [
    ["name","age"],
    ["raj",12],
    ["rajesh",12],
    ["raju",12],
    ["rajendra",12],
    ["rakesh",12],
]
# data = [
#     "raj",
#     "rajesh"
# ]
```

```python
with open("first_csv.csv","w+") as file:
    writer = csv.writer(file,)
    for row in data:
        writer.writerow(row)
```

```python
try :
    with open("first_csv.csv","x+") as file:
        writer = csv.writer(file,)
        for row in data:
            writer.writerow(row)
except Exception as e:
    print("Exception",e)
except FileExistsError as e:
    with open("first_csv2.csv","x+") as file:
        writer = csv.writer(file,)
        for row in data:
            writer.writerow(row)
    print("FileExistsError",e)
except ZeroDivisionError as e:
    print("ZeroDivisionError",e)
except IndexError as e:
    print("IndexError",e)

print("done")
```
