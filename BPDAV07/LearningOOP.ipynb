# APIE – OOP Concepts

## 1. Abstraction
- Abstraction in OOP is a design principle that involves **hiding complex implementation details** and showing only the **essential features** of an object or system.
- It helps in **reducing complexity** and **providing a simplified interface**.
- Abstraction is the process of extracting **important information** about the domain and ignoring irrelevant details.

---

## 2. Polymorphism
- **Poly** = many, **morph** = forms → *many forms*.
- Polymorphism allows the same function or method to behave **differently** in different contexts.

### Types of Polymorphism

#### 1. Overloading
- Overloading is the process of redefining the:
<pre>
same method
with the same name
within the SAME class
but with different sets of parameters.
</pre>

- **Python does NOT support method overloading** directly.

#### 2. Overriding
- Overriding is the process of redefining the:
<pre>
same method
with the same name
and SAME parameters
within the CHILD class.



---

## 3. Inheritance
- Inheritance allows one class (**child/derived class**) to acquire the attributes and methods of another class (**parent/base class**).
- Promotes **code reusability**, **extensibility**, and **hierarchical organization**.

### Types of Inheritance
- **Single Inheritance**
- **Multilevel Inheritance**
- **Multiple Inheritance**
- **Hierarchical Inheritance**
- **Hybrid Inheritance**

---

## 4. Encapsulation
- Encapsulation is the process of **wrapping data (variables)** and **methods** into a single unit (class).
- It **restricts direct access** to some data fields.
- Ensures:
  - **Data protection**
  - **Security**
  - **Controlled access** using getters and setters

Python uses:
- `_protectedVariable`
- `__privateVariable`

to indicate restricted access.


---
## Abstract Classes
* Abstract classes are the classes which hold abstract mehtod
* Abstract class can't be instantiate (can't create object of abstract class)
* body of abstract method has to be given by the child of abstract class 

## Abstract Method
* Abstract method are the method which hold only declaration not the definition (no body)
* the defination of abstract mehtod should be given by the child class by overriding method

```python
import abc
from abc import abstractmethod
class Dominos(abc.ABC):
    @abstractmethod
    def online_order(self):
        pass
    @abstractmethod
    def delievery(self):
        pass
    @abstractmethod
    def pick_up(self):
        pass

class MyRestro(Dominos):
    def online_order(self):
        return super().online_order()
    def pick_up(self):
        return super().pick_up()
    def delievery(self):
        return super().delievery()
borivali = MyRestro()
```

```python
import abc
from abc import abstractmethod
class PersonTemplate(abc.ABC):
    @abstractmethod
    def do_work(self):
        pass
class Person(PersonTemplate):
    def __init__(self,name,age,gender,height,weight):
        self.name = name
        self.age = age
        self.gender = gender
        self.height = height
        self.weight = weight
    def do_work(self):
        return super().do_work()
p1 = Person("Raj",22,"Male","5f2in","60kg")
p1.__dict__
```

## Constructor
- This is a **special method** in Python classes that gets automatically called when a new object is created from the class.
- Its primary purpose is to initialize the attributes of the newly created object.
- It always takes self as its first argument, which refers to the instance of the object being created. Subsequent arguments are used to set initial values for the object's attributes.


---
### Types of Method
* instance method where you can find self first parameter
* class method where you can find cls as a first parameter
* **to create class method `@classmethod decorator` is requied**
* static method with no specified parameter as a first
* **to create static method `@staticmethod decorator` is requied**


```python
class Student:
    school_name = "Xaviers"     # class level attribut
    # constructor 
    # 1. object return
    # 2. assign the value to the instance at time of creation
    def __init__(self,name,age):
        self.name = name
        self.age = age
    @classmethod
    def set_school_name(cls,new_name):
        cls.school_name = new_name
    def learn(self,skill):
        print(self.name+" Learning ",skill)
        return True
```
