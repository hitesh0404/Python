* Use short, all-lowercase names.
* Use underscores for multiple words (e.g., my_module.py, data_processing).

### Classes:
* Use CapWords (CamelCase), capitalizing the first letter of each word without underscores (e.g., MyClass, HTTPRequest).
### Exception classes should end with "Error" (e.g., ValueError).
### Functions and Variables (including global and instance variables):
* Use lowercase with underscores (snake\_case) for multiple words (e.g., my_function, total_count).
### Constants:
* Use ALL_CAPS with underscores for multiple words (e.g., MAX_RETRIES, DEFAULT_TIMEOUT).
### Methods:
* Follow the same conventions as functions (snake\_case).
### Method Arguments:
* Follow the same conventions as variables (snake\_case).
### Private Members (by convention):
* Prefix with a single underscore (e.g., _private_attribute, _private_method). This indicates to other developers that these are internal and not intended for direct external use.
### Name Mangling (strongly private):
* Prefix with double underscores (e.g., __mangled_attribute). This causes the interpreter to "mangle" the name, making it harder to access directly from outside the class.
### Special Methods ("Dunder Methods"):
* Prefix and suffix with double underscores (e.g., __init__, __str__). These are reserved for built-in methods with special meaning in Pyth"""

```python
class Student:
    college = "ACV"
    def __init__(self,name,age):   # constructor
        self.name = name
        self.age = age
    def get_college_name(cls):
        return cls.college
    def intro(self):
        return f"hello my name is {self.name} and i am { self.age} years old"

raj = Student("Raj Rathod",23)
rajesh = Student("Rajesh Rane",32)

print(raj.intro())
print(rajesh.intro())
print(raj.get_college_name())
print(rajesh.get_college_name())
```

```python
raj.name = "Raj Rathod"
raj.age = 12
rajesh.name = "Rajesh Rane"
rajesh.age = 14
```

```python
from datetime import datetime
from datetime import date, timedelta
class GYM_Members:
    def __init__(self,name,address,schedule,start_date):
        self.name = name
        self.address = address
        self.schedule = schedule
        self.start_date = start_date
    def buy_membership(self,duration,amount):
        self.duration = duration
        self.amount = amount
    def extend_membership(self,days,amount):
        self.duration += days
        self.amount += amount
    def membership_validity(self):
        try :
            format2 = r"%d-%m-%Y"
            return (datetime.strptime(self.start_date, format2) + timedelta(self.duration))
        except  ValueError as e:
            print(f"Error: {e}")


hitesh = GYM_Members("Hitesh","vasai","5 to 7","07-10-2025")
hitesh.buy_membership(60,3000)
hitesh.extend_membership(60,3000)


hitesh.membership_validity()
```
