```python
def get_primes(start,stop):
    result =[]
    total_prime=0
    for num in range(start,stop):
        prime = True
        for i in range(2,num):
            if num%i==0:
                prime = False
                break
        if prime:
            result.append(num)
            total_prime+=1
    return result,total_prime        
```

```python
get_primes(20,30)
```

```python
class Singleton:
    _instance = None # falsy value. truth

    def __new__(cls, *args, **kwargs):

        if not cls._instance:
            cls._instance = super().__new__(cls)
            print("first time",cls._instance)
        return cls._instance

    def __init__(self, value):
        if not hasattr(self, '_initialized'):  # Prevent re-initialization
            self.value = value
            self._initialized = True

# Usage
s1 = Singleton("First instance")
print(s1.value)

s2 = Singleton("Second instance") # This will return the same instance as s1
print(s2.value)

print(s1 is s2)
```

```python
def singleton(cls):
    instances = {}                                      #dictionary
    # {class_name: instance}
    # {MyClass : obj}
    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return get_instance

@singleton
class MyClass:
    def __init__(self, name):
        self.name = name
from math import pi
@singleton
class MathFunction:
    def __init__(self):
        self.pi = pi
    def perimeter_of_circle(self,radius):
        return f"{2 * self.pi * radius:.2f}"
    def area_of_circle(self,radius):
        return f"{self.pi*radius**2:.2f}"

obj1 = MathFunction()
print(obj1.perimeter_of_circle(32))
print(obj1.area_of_circle(32))

# Usage
m1 = MyClass("Alice")
print(m1.name)

m2 = MyClass("Bob") # This will return the same instance as m1
print(m2.name)

print(m1 is m2)
```
