## pd.Series
- one dimentional collection of element of similar data type
- similar to Numpy array
- dedicated index
- support different pandas methods


```python
import pandas as pd

s = pd.Series([1,2,3,4,1],index=["a","b","c","d","e"])

print(f"mean  {s.mean()}".rjust(30," "))
print(f"standard deviation  {s.std():.2f}".rjust(30," "))
print(f"sum  {s.sum():.2f}".rjust(30," "))
print(f"mode  {s.mode()}".rjust(30," "))
```

## Step 1: Null Value 

```python
selected = df1[df1["normalizedLosses"]=="?"].index

df1.loc[selected,"normalizedLosses"] = "0"

# df1["normalized-losses"] = df1["normalized-losses"].replace("?",0).astype(int)
df1["normalizedLosses"]=df1["normalizedLosses"].astype("int")

# df["normalized-losses"] = df["normalized-losses"].replace(0,df["normalized-losses"].median())
df1.loc[selected,"normalizedLosses"] = df1["normalizedLosses"].median()
```

```python
df1.head()
```

```python
df1["horsepower"] = df1["horsepower"].replace("?",0).astype(int)
df1["horsepower"] = df1["horsepower"].replace(0,df1["horsepower"].median())
```

## iter

- which can be used to iterate over iterable with next function

```python
it = iter(["a","b","c","d"])
```

```python
next(it)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = df1.select_dtypes(exclude=object).columns
it = iter(cols)

fig,ax = plt.subplots(3,3,figsize=(20,10))
for i in range(3):        # 0              1              2
    for j in range(3):    # 0 1 2          0 1 2          0 1 2
        col = next(it)
        sns.boxplot(df1,x = col,ax=ax[i,j]) 
        ax[i,j].set_title(col)
plt.tight_layout()
plt.show()
```

```python
df1.columns
```

```python
outliers = set()
d = list(df1[df1["normalizedLosses"]>190].index)
outliers.update(d)
d = list(df1[df1["width"]>71.5].index) 
outliers.update(d)
d = list(df1[df1["engineSize"]>210].index) 
outliers.update(d)
d = list(df1[df1["horsepower"]>190].index) 
outliers.update(d)
d = list(df1[df1["cityMpg"]>45].index) 
outliers.update(d)
d = list(df1[df1["highwayMpg"]>47].index) 
outliers.update(d)
d = list(df1[df1["price"]>30000].index) 
outliers.update(d)
len(outliers)
```

```python
df1.drop(outliers,inplace=True)
```

# 4. Encoding Categorical Data(transfering category into number)
- 1. Label Encoder(sklearn.preprocessing.LabelEncoder)
- 2. One Hot Encoder(pd.get_dummies)

```python
from sklearn.preprocessing import LabelEncoder

le_objects = {}

for col in df1.select_dtypes(include=object).columns:
    le = LabelEncoder()
    df1[col] = le.fit_transform(df1[col])
    le_objects[col] = le
```
