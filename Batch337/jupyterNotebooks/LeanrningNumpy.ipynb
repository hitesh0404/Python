```python
np.random.seed(42)
data = np.random.randint(1,11,100)
print(data.size)
data.sort()


mid_index = int(data.size/2)-1
if data.size %2==0:                        # if even count
    values = data[mid_index:mid_index+2]    # [49:51]  these are index or position
    print(values.sum()/2)
else:                                      # if odd count
    values = data[mid_index:mid_index+1] 
    print(values)

print(np.quantile(data,0.5))
```

```python
import numpy as np

hours = np.array([1, 2, 3, 4])

marks = np.array([40, 50, 65, 80])

cov_matrix = np.cov(hours, marks, bias=True) # if true population divided by n
                                             # if false sample divided by n-1 

print("Covariance:", cov_matrix[0][1])        
print("Correlation =" , cov_matrix[0][1] / (hours.std()*marks.std()) )   
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.DataFrame({"hours":hours,"marks":marks})
plt.figure(figsize=(6,4))
sns.scatterplot(df,x="hours",y="marks")
sns.lineplot(df,x="hours",y="marks")
```

```python
# to remove skewness
np.random.seed(42)
data = np.random.randint(1,100,1000)
data = np.hstack((data,np.random.randint(1,30,1000)))
df = pd.DataFrame({"col1":data})
df["col1"] = np.sin(df["col1"])    # log1p sqrt  sin cos
plt.figure(figsize=(6,2))
sns.histplot(df,bins=30)
df["col1"].skew()
```
