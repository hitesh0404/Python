```python
import pandas as pd
from faker import Faker
import random

random.seed(42)
Faker.seed(42)
# Initialize the Faker object
fake = Faker("en_IN")

# Define the number of fake records you want to generate
num_records = 100

# Generate the dataset
data = []

for _ in range(num_records):
    game_title = fake.catch_phrase()  # Fake game title
    genre = random.choice(['Action', 'Adventure', 'RPG', 'Shooter', 'Sports', 'Racing', 'Puzzle', 'Simulation'])  # Random genre
    publisher = fake.company()  # Fake company name (publisher)
    year = random.randint(2013, 2023)  # Random year between 2013 (PS4 launch year) and 2023
    global_sales = round(random.uniform(0.1, 50.0), 2)  # Random global sales between 0.1 million to 50 million

    data.append([game_title, genre, publisher, year, global_sales])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Name', 'Genre', 'Publisher', 'Year', 'Global_Sales'])

# Show the generated DataFrame
print(df.head())

# Optionally, save it to a CSV file
df.to_csv('fake_ps4_sales_data.csv', index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("fake_ps4_sales_data.csv")
```

# Bar plot

```python
df.info()
```

```python
df_bar = df.groupby(by="Year")["Global_Sales"].sum()
```

```python
plt.bar(df_bar.index,df_bar.values)
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.title("Global Sales yearly")
plt.xticks(rotation=90) #
plt.show()
```

```python
sns.barplot(data=df,x="Year",y="Global_Sales",estimator="sum",orient="v")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.title("Global Sales yearly")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.barh(df_bar.index,df_bar.values)
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.title("Global Sales yearly")
plt.xticks(rotation=90) #
plt.show()
```

```python
sns.barplot(data=df,x="Global_Sales",y="Year",estimator="sum",orient="h")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.title("Global Sales yearly")
plt.xticks(rotation=90)
plt.show()
```
