# Pandas
## Datatypes
### 1. Series
- just like one dimentional array of numpy 
- but it support wide veriety of pandas method
- it support dedicated index
- and easy to work with DataFrame
### 2. DataFrame
- just like two dimentional(table) array of numpy 
- it support dedicated index
- it support wide veriety of pandas method
- Pandas supports multi-value (hierarchical or multi-level) headers through the MultiIndex object

```python
s = pd.Series([1,2,3,3,4,5,6,7,8,9,10],list("abcdefghijk"))
s["a":"d"]
```

```python
print(s.mean())
print(s.median())
print(s.mode())
```

```python
print(s.apply(lambda x: x*2))
```
