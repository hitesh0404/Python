## Set 
- identified by {}
- value should be hashable
- doesn't allow non hashable object (imutable objects  list,set and dictinary)
- it is non sequencial object
- only unique elements (no duplicate elements)

```python
s = {12,3,(1,2,3),1,2,1,2,1,2,1,2}
s
```

```python
#remove duplicates
data = [12,31,31,12,13,1]
i=0
while i < len(data):
    if data.count(data[i])>1:
        data.remove(data[i])
    i+=1
print(data)
```

```python
data = [12,21,21,12,13,1]
data = list(set(data))
print(data)
```

```python
data = [12,31,31,12,13,1]
i=0
while i < len(data):
    if data.count(data[i])>1:
        data.pop(data.index(data[i],i+1))
    i+=1
print(data)
```
