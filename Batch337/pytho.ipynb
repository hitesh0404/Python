```python
with open("hello.txt","a+") as file: # pointer at end
    print("writable",file.writable())
    print("readable",file.readable())
    file.seek(10,0)
    # cookie: 10,           count of step
    # whence: int = 0        0 = start    1 = current    2 = end
    print(file.read())
```

## recursion
- when function call it self.

```python
def factorial(num):  # 4
    if num<=1:
        return 1
    else:
        print(num)
        return num * factorial(num-1)
```

```python
factorial(4)
```

## nested Functions

- when function create a inner function
- clouser

```python
def letsStart(num):
    count = num
    def countSomething(increaseBy):
        nonlocal count
        count += increaseBy
        return count
    return countSomething
```

```python
player1 = letsStart(3)
player2 = letsStart(4)
```

```python
player1(0)
```

```python
player2(0)
```

```python
d = [1,2,3,4,5]
for i,v in enumerate(d):
    d[i] = v*2
print(d)

def double(x):
    return x*2

list(map(double,d))
```

## anonymous Function
- lambda function
- single line function
- it can take n parameters
- does not require to return value

```python
dub = (lambda x,y,z : x+y+z)
dub(23,1,1)
```

```python
list(map(lambda x : x*2 , [1,2,3,4]))
```

```python
def my_map(fun,obj):
    final = []
    for i in obj :
        final.append(fun(i))
    return final

list(my_map(lambda x : x*2 , [1,2,3,4]))
```

```python
def my_map(fun,obj):
    for i in obj :
        yield fun(i)

print(list(my_map(lambda x : x*2 , [1,2,3,4])))
```

### method types

### instance level method
- self is first parameter
- it can directly access the current instance
- it can only be called by object(instance) of class

```python
class Fruit:
    flavour = "sweet"         # class level attribute
    def __init__(self,name):
        self.name = name
    def intro(self):
        return "my name is " + self.name
    @classmethod       
    def description(cls):
        return "this is class " + cls.__name__
    @staticmethod
    def doSomething():
        return "anything means nothing"

date = Fruit("Date")
date.intro()
date.description()
Fruit.description()
Fruit.doSomething()
date.doSomething()
```
